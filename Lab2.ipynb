```python
import pandas as pd
import numpy as np
prices = pd.read_csv("data/sample_prices.csv")
returns = prices.pct_change()
returns
```

```python
returns = returns.dropna()
returns
```

```python
deviations = returns - returns.mean()
square_deviations = deviations**2
variance = square_deviations.mean()
std = np.sqrt(variance)
std
```

```python
no_of_obs = returns.shape[0]
no_of_obs
```

```python
variance = square_deviations.sum()/(no_of_obs-1)
std = np.sqrt(variance)
std
```

```python
annualized_vol = std*np.sqrt(12)
annualized_vol
```

```python
me_m = pd.read_csv("data/Portfolios_Formed_on_ME_monthly_EW.csv",
                  header = 0, index_col=0, parse_dates=True, na_values=-99.99)
me_m.head()
```

```python
columns = ['Lo 10','Hi 10']
returns = me_m[columns]
returns.head()
```

```python
returns = returns/100
```

```python
returns
```

```python
returns.plot.line()
```

```python
annualized_vol = returns.std()*np.sqrt(12)
annualized_vol
```

```python
n_months = returns.shape[0]
return_per_month = (returns + 1).prod()**(1/n_months) - 1
return_per_month
```

```python
annualized_return = (return_per_month + 1)**12 - 1
annualized_return
```

```python
annualized_return= (returns + 1).prod()**(12/n_months) -1
annualized_return
```

```python
riskfree_rate = 0.03
excess_returns = annualized_return - riskfree_rate
sharpe_ratio = excess_returns/annualized_vol
sharpe_ratio
```
